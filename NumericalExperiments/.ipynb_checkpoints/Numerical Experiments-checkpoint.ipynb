# IMAGE SEPERATION EXAMPLE (BATCH LEARNING)

```python
I = np.load('test_images.npy') ## I contains 4 images. We will illustrate image seperation examples for both batch and online learning of the proposed model.
```

```python
S = I[:,0:3]#Generate a source matrix from 3 images
print('Below are the original source images')
Subplot_gray_images(S)
```

```python
A = np.random.randn(4,3)  # Uniform Mixing matrix
X = np.dot(S, A.T)  # Generate observations
print('Below are the mixture signals')
Subplot_gray_images(X)
```

```python
# HYPERPARAMETERS DEFINED IN THE PAPER
lambda_ = 0.999
mu_F = 0.03
beta = 5

model1 = OnlineBCA(s_dim = 3, x_dim = 4, lambda_= lambda_, mu_F = mu_F, beta = beta, neural_dynamic_iterations = 100)
Y1 = model1.fit_batch(X, n_epochs = 1, shuffle = True, verbose = True)
```

There is a sign and permutation ambiguity in the seperated sources, as it also happens in ICA. Therefore, we imshow both the seperated sources and the negative of the seperated sources

```python
print('Below are the seperated sources')
Subplot_gray_images(ZeroOneNormalizeColumns(Y1))
```

```python
print('Below are the negative of the seperated sources')
Subplot_gray_images(ZeroOneNormalizeColumns(-Y1))
```

# IMAGE SEPERATION (ONLINE LEARNING)

```python
A = np.random.randn(3,3)  # Uniform Mixing matrix
X = np.dot(S, A.T)  # Generate observations

s_dim = S.shape[1]
x_dim = X.shape[1]
samples = S.shape[0]
print('Below are the mixture signals (after whitening)')
X = whiten_signal(X)[0]
Subplot_gray_images(X)
```

```python
model2 = OnlineBCA(s_dim = s_dim, x_dim = x_dim, lambda_= lambda_, mu_F = mu_F, beta = beta, neural_dynamic_iterations = 100)
Y2 = np.zeros((samples,s_dim))
idx = np.random.permutation(samples) # random permutation

for i_sample in tqdm(range(samples)):
    x_current = X[idx[i_sample],:]
    Y2[idx[i_sample],:] = model2.fit_next(x_current)
```

```python
print('Below are the seperated sources')
Subplot_gray_images(ZeroOneNormalizeColumns(Y2))
```

```python
print('Below are the negative of the seperated sources')
Subplot_gray_images(ZeroOneNormalizeColumns(-Y2))
```
